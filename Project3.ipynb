# Promedio
--------
> La siguiente tabla presenta el salario promedio de los países elegidos para analizar en los dos sectores seleccionados.

---------


```python
import pandas as pd

file_path = '/content/sample_data/1. DATOS/global_jobs_salaries_2024.csv'
df = pd.read_csv(file_path)

countries = ['Chile', 'Mexico', 'Singapore', 'Uruguay']
categories = ['Factory and Manufacturing', 'Oil / Gas / Energy / Mining']
filtered_df = df[(df['Country'].isin(countries)) & (df['Category'].isin(categories))]

average_salaries = filtered_df.groupby(['Country', 'Category'])['Salary (USD)'].mean().reset_index()

pivot_table = average_salaries.pivot(index='Country', columns='Category', values='Salary (USD)')

pivot_table = pivot_table.round(2)

print(pivot_table)
```

A continuación se presenta una gráfica de barras que muestra el salario promedio en los dos sectores seleccionados, "Fábrica y Manufactura" (Factory and Manufacturing) y "Petróleo / Gas / Energía / Minería" (Oil / Gas / Energy / Mining), en cuatro países: Chile, México, Singapur y Uruguay. Es útil para entender cuánto se está pagando, en promedio, en cada país y sector.

```python
import matplotlib.pyplot as plt
import numpy as np

categorias = ['Fábrica y Manufactura', 'Petróleo / Gas / Energía / Minería']
uruguay_salarios = [7357.51, 9645.29]
chile_salarios = [15660.24, 20634.96]
mexico_salarios = [16609.14, 21881.29]
singapur_salarios = [49070.70, 64451.53]

x = np.arange(len(categorias))
width = 0.18

fig, ax = plt.subplots(figsize=(10, 6))

bars_uruguay = ax.bar(x - width * 1.5, uruguay_salarios, width, label='Uruguay', color='#A4C8E1')
bars_chile = ax.bar(x - width / 2, chile_salarios, width, label='Chile', color='#F4C2C2')
bars_mexico = ax.bar(x + width / 2, mexico_salarios, width, label='México', color='#A8D5BA')
bars_singapur = ax.bar(x + width * 1.5, singapur_salarios, width, label='Singapur', color='#C4A7E7')

for i, bar in enumerate(bars_uruguay):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height()/2, f'{uruguay_salarios[i]:,.0f}', ha='center', va='center', color='navy')

for i, bar in enumerate(bars_chile):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height()/2, f'{chile_salarios[i]:,.0f}', ha='center', va='center', color='deeppink')

for i, bar in enumerate(bars_mexico):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height()/2, f'{mexico_salarios[i]:,.0f}', ha='center', va='center', color='green')

for i, bar in enumerate(bars_singapur):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height()/2, f'{singapur_salarios[i]:,.0f}', ha='center', va='center', color='purple')

ax.set_xlabel('Trabajos')
ax.set_ylabel('Salario promedio (miles USD)')
ax.set_title('Salario promedio por país y categoría (anual 2024)', fontsize=15)
ax.set_xticks(x)
ax.set_xticklabels(categorias)
ax.legend(loc='upper left', bbox_to_anchor=(0.02, 0.98), title="Referencias: ", fontsize=10)

ax.set_yticks([20000, 40000, 60000])
ax.set_ylim(0, 65000)
ax.set_yticklabels(['20', '40', '60'])

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.text(0.5, -0.12, 'Fuente: Global Jobs and Salaries 2024', ha='center', va='center', fontsize=9, color='gray', transform=ax.transAxes)

plt.xticks(rotation=0, ha="center")

plt.tight_layout()

plt.show()
```

# Tabla de valores

La tabla muestra los salarios más altos y bajos en Chile, México, Singapur y Uruguay, divididas en los sectores de trabajos por Fábrica y Manufactura y Petróleo / Gas / Energía / Minería.

```python

import pandas as pd

file_path = '/content/sample_data/1. DATOS/global_jobs_salaries_2024.csv'
df = pd.read_csv(file_path)

countries = ['Chile', 'Mexico', 'Singapore', 'Uruguay']
categories = ['Factory and Manufacturing', 'Oil / Gas / Energy / Mining']
filtered_df = df[(df['Country'].isin(countries)) & (df['Category'].isin(categories))]

salary_stats = filtered_df.groupby(['Country', 'Category'])['Salary (USD)'].agg(['max', 'min']).reset_index()

pivot_max = salary_stats.pivot(index='Country', columns='Category', values='max').rename(columns=lambda x: f"{x} (Max)")
pivot_min = salary_stats.pivot(index='Country', columns='Category', values='min').rename(columns=lambda x: f"{x} (Min)")

result_table = pd.concat([pivot_max, pivot_min], axis=1)

result_table = result_table.round(2)

print(result_table)
```

> En base a la tabla presentada anteriormente, se crea una gráfica sobre los salarios más altos, más bajos y el promedio en los dos sectores de los cuatro países seleccionados. Permite comparar las diferencias salariales en cada país, identificar desigualdades entre los salarios más bajos y los más altos, y analizar cómo se posiciona el promedio en cada sector.

# Valores máximos y mínimos

> Datos:


Categoría:  Fábrica y Manufactura (Factory and Manufacturing)

(Max)

* Chile (Salary USD): 40426.31                             
* México (Salary USD): 42969.86                             
* Singapur (Salary USD): 117578.18                             
* Uruguay (Salary USD): 19215.91    

(Mín)

* Chile (Salary USD): 5599.86
* México (Salary USD): 6289.89
* Singapur (Salary USD): 17100.94
* Uruguay (Salary USD): 2715.15   

----------------------------------

Categoría: Petróleo / Gas / Energía / Minería (Oil / Gas / Energy / Mining)

(Max)

* Chile (Salary USD): 49921.83   
* México (Salary USD): 49861.65   
* Singapur (Salary USD): 147804.12   
* Uruguay (Salary USD): 21583.28

(Mín)

* Chile (Salary USD): 7840.33  
* México (Salary USD): 7734.45  
* Singapur (Salary USD): 23160.91  
* Uruguay (Salary USD): 3347.34  


```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

data = {
    'Country': ['Uruguay', 'Chile', 'México', 'Singapur'],
    'Min': [2715.15, 5599.86, 6289.89, 17100.94],
    'Promedio': [7357.51, 15660.24, 16609.14, 49070.70],
    'Max': [19215.91, 40426.31, 42969.86, 117578.18]
}

df = pd.DataFrame(data)

pastel_colors = {
    'min': '#F6E58D',
    'mean': '#FF9F43',
    'max': '#FF6F61',
    'line': '#D1D1D1'
}

def abbreviate_number(x, pos):
    if x >= 1e6:
        return f'{x*1e-6:.0f}M' # Millones
    elif x >= 1e3:
        return f'{x*1e-3:.0f}K' # Miles
    else:
        return f'{int(x)}'  # Sin abreviación si es < mil

fig, ax = plt.subplots(figsize=(10, 6))

countries = df['Country'].tolist()

for i, country in enumerate(countries):
    min_salary = df[df['Country'] == country]['Min'].values[0]
    mean_salary = df[df['Country'] == country]['Promedio'].values[0]
    max_salary = df[df['Country'] == country]['Max'].values[0]

    ax.plot([country, country], [min_salary, max_salary], color=pastel_colors['line'], linestyle='--', alpha=0.5)

    ax.plot(country, min_salary, marker='o', color=pastel_colors['min'], markersize=10)
    ax.text(country, min_salary + 2000, f'{min_salary:,.0f}', color='grey', ha='center', va='center', fontsize=8, fontweight='bold')

    ax.plot(country, mean_salary, marker='o', color=pastel_colors['mean'], markersize=10)
    ax.text(country, mean_salary + 5500, f'{mean_salary:,.0f}', color='orange', ha='center', va='center', fontsize=8, fontweight='bold')

    ax.plot(country, max_salary, marker='o', color=pastel_colors['max'], markersize=13)
    ax.text(country, max_salary + 5500, f'{max_salary:,.0f}', color='red', ha='center', va='center', fontsize=8, fontweight='bold')

ax.set_title('Salario en Fábrica y Manufactura por países anual (2024)', fontsize=13, ha="center")
ax.set_xlabel('Países', fontsize=10)
ax.set_ylabel('Salario (miles USD)', fontsize=10)

ax.set_xticks(countries)
ax.set_xticklabels(countries, rotation=0, ha="center")

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.yaxis.set_major_formatter(FuncFormatter(abbreviate_number))

ax.plot([], [], color=pastel_colors['min'], label='Salario Mínimo', marker='o', markersize=7)
ax.plot([], [], color=pastel_colors['mean'], label='Salario Promedio', marker='o', markersize=7)
ax.plot([], [], color=pastel_colors['max'], label='Salario Máximo', marker='o', markersize=7)

ax.legend(title="Referencias: ", loc='upper left', bbox_to_anchor=(0.05, 1), frameon=True)

plt.text(0.5, -0.12, 'Fuente: Global Jobs and Salaries 2024', ha='center', va='center', fontsize=9, color='gray', transform=ax.transAxes)

plt.tight_layout()
plt.show()
```
